```python
# Strip whitespaces in Column headers
df_books.columns = df_books.columns.str.replace(' ', '_')
df_customers.columns = df_customers.columns.str.replace(' ', '_')
print(df_books.head())
print(df_customers.head())
```

```python
# Drop Rows from file with missing data
cl_dfbooks = df_books.dropna(axis=0)
cl_df_customers = df_customers.dropna(axis=0)
```
